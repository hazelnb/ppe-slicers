```python
from slicers import BandSlicer
import shapely
import geopandas as gpd
```

```python
il_gdf = gpd.read_file("data/tl_2020_17_county20/tl_2020_17_county20.shp")
il = il_gdf.dissolve().to_crs("epsg:3435")

square = shapely.Polygon([[0,0],[0,1],[1,1],[1,0]])
circle = shapely.Point((0,0)).buffer(1)
```

```python
il_bs = BandSlicer(il, tolerance=0.001, n_bands=8, band_width=.034, piece_width=.031) #units are meters
il_bs.plot()
```

```python
sq_bs = BandSlicer(square, tolerance=0, n_bands = 10, band_width = 0.053,piece_width = 0.053)
sq_bs.plot()
```

```python
circ_bs =  BandSlicer(circle, tolerance=0, n_bands = 9, band_width = 0.058, piece_width = 0.058)
circ_bs.plot()
```
